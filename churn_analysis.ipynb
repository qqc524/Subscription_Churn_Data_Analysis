## **Categorical Feature Analysis**

```python
categorical_columns = categorical_columns = ['SubscriptionType', 'PaymentMethod', 'PaperlessBilling', 'ContentType', 
                       'MultiDeviceAccess', 'DeviceRegistered', 'GenrePreference', 'Gender', 
                       'ParentalControl', 'SubtitlesEnabled']
```

### Feature Distributions

```python
# Visualize the categorical feature distributions
num_columns = len(categorical_columns)
rows = (num_columns // 2) + (num_columns % 2)  
cols = 2

fig, axes = plt.subplots(rows, cols, figsize=(15, rows * 5))  # Adjust figure size based on rows
axes = axes.flatten() if isinstance(axes, np.ndarray) else [axes]  # Ensure axes is iterable

# Plot each categorical feature
for i, col in enumerate(categorical_columns):
    churn[col].value_counts().plot(kind='bar', color= '#f6ddcc', edgecolor='black', ax=axes[i])
    axes[i].set_title(f'{col} Distribution', fontsize=14)
    axes[i].set_xlabel(col, fontsize=12)
    axes[i].set_ylabel('Count', fontsize=12)
    axes[i].tick_params(axis='x', rotation=45)


# Turn off unused subplot axes
for i in range(num_columns, len(axes)):
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

### Categorical Features vs Churn

<span style="font-size:16px">
The analysis explores how each categorical feature relates to the target variable, Churn. 
Bar plots are used to visualize the distribution of each category within a feature, comparing the count of users who churned (1) versus those who did not (0). These insights help identify patterns and relationships between customer attributes and churn behavior.
</span>

```python
# Bar plots for categorical features
for column in categorical_columns:
    plt.figure(figsize=(10, 6))
    ax = sns.countplot(x=column, hue='Churn', data=churn, palette=colors)
    
    # Add count annotations
    for container in ax.containers:
        ax.bar_label(container, fmt='%d', label_type='edge', fontsize=10)
    
    plt.title(f'{column} vs Churn', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Count', fontsize=12)
    
    # Move the legend outside the plot
    plt.legend(title='Churn', labels=labels, fontsize=12, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.xticks(rotation=45)
    plt.tight_layout()  # Ensure the layout is tight to prevent clipping
    plt.show()
```

### Boxplot of Features Distribution 

<span style="font-size:16px;">
A boxplot is a graphical representation of the distribution of a dataset. It is used to visually summarize the spread, center, and outliers in the data.
</span>

#### Why We Use a Boxplot
Boxplots are useful for:
- **Detecting outliers**: Data points outside the whiskers of the box are considered outliers.
- **Understanding the spread** of the data: It shows the range and variation of the data.
- **Comparing distributions**: It helps compare data from different categories or groups.


#### The Five-Number Summary

A boxplot is made using a **five-number summary**, which summarizes the dataset in a simple way:

- **Minimum**: The smallest value in the dataset.
- **First Quartile (Q1)**: The middle value of the lower half of the data. It represents the **25th percentile**.
- **Median (Q2)**: The middle value that divides the data into two equal halves.
- **Third Quartile (Q3)**: The middle value of the upper half of the data. It represents the **75th percentile**.
- **Maximum**: The largest value in the dataset.



```python
# Loop through each numeric column
for i, column in enumerate(numerical_columns):
    # Create a boxplot for each column
    plt.figure(figsize=(8, 6))
    
    # Plot boxplot
    ax = sns.boxplot(y=churn[column], color='lightcoral')

    # Title and labels
    plt.title(f'Boxplot of {column}', fontsize=14)
    plt.xlabel(column, fontsize=12)
    plt.ylabel('Value', fontsize=12)

    # Add details for the 5-number summary and outliers
    Q1 = churn[column].quantile(0.25)
    Q3 = churn[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = churn[column][(churn[column] < lower_bound) | (churn[column] > upper_bound)]

    # Annotate outliers in the plot
    if len(outliers) > 0:
        plt.scatter([0] * len(outliers), outliers, color='blue', label='Outliers')

    # Add 5-number summary and outlier count text outside the graph
    text_str = (f"Q1: {Q1:.2f}\nQ3: {Q3:.2f}\nIQR: {IQR:.2f}\n"
                f"Minimum: {lower_bound:.2f}\nMaximum: {upper_bound:.2f}\n"
                f"# of Outliers: {len(outliers)}")
    
    plt.text(1.05, 0.5, text_str, transform=ax.transAxes, fontsize=12,
             verticalalignment='center', horizontalalignment='left', bbox=dict(facecolor='white', alpha=0.8))

    # Show the plot
    plt.tight_layout()
    plt.show()
```

## **Correlation Heatmap for Numeric Variables**


Correlation Coefficient:

- 1: Perfect positive correlation
- 0: No correlation
- -1: Perfect negative correlation

```python
# Calculate correlation matrix for numeric variables
correlation_matrix = churn.corr(numeric_only=True)


# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True, linewidths=0.5)
plt.title("Correlation Heatmap for Numeric Variables", fontsize=14)
plt.xticks(fontsize=10, rotation=65)
plt.yticks(fontsize=10)

plt.tight_layout()
plt.show()
```

<span style="font-size:16px;">
The correlation heatmap visually represents the relationships between numerical variables in the dataset. Most correlations are close to 0, indicating weak or no linear relationships among the variables. This suggests that the numerical features are generally independent of one another.
</span>

## **Feature Interaction with Churn**

### Taking a closer look at the correlation between Churn and other variables:


```python
# Calculate correlation with Churn
correlation_with_churn = churn.corr(numeric_only=True)['Churn'].sort_values()

# Plot the correlation as a bar chart
fig, ax = plt.subplots(figsize=(10, 8))
bars = ax.bar(
    correlation_with_churn.drop('Churn').index, 
    correlation_with_churn.drop('Churn'), 
    color='lightcoral', 
    edgecolor='black'
)

# Add annotations for correlation rates
ax.bar_label(bars, fmt='%.2f', fontsize=10, label_type='edge')

# Add titles and labels
ax.set_title('Correlation of Numeric Variables with Churn', fontsize=14)
ax.set_xlabel('Features', fontsize=12)
ax.set_ylabel('Correlation Coefficient', fontsize=12)
ax.tick_params(axis='x', labelrotation=45, labelsize=10) 
plt.tight_layout()  
plt.show()
```

### Insights:

- The weak correlations suggest that no single numeric variable has a strong direct relationship with Churn.
- **AccountAge, ViewingHoursPerWeek, and AverageViewingDuration** show slight negative correlations with Churn, indicating that longer engagement may reduce churn risk.
- **MonthlyCharges and SupportTicketsPerMonth** show slight positive correlations, suggesting that higher charges or frequent support issues might contribute to churn.

### <u>AccountAge vs Churn:</u>

**AccountAge (-0.20):** A weak negative correlation suggests that customers with longer account ages might be less likely to churn. 

```python
plt.figure(figsize=(8, 6))
ax = sns.boxplot(x='Churn', y='AccountAge', data=churn, hue='Churn', palette=colors, legend=False)

# Calculate the median values for both groups (Churned and Not Churned)
medians = churn.groupby('Churn')['AccountAge'].median()

# Annotate the medians on the boxplot
for i, median in enumerate(medians):
    ax.text(i, median, f'{median:.2f}', horizontalalignment='center', size=12, color='black', fontweight='bold')

# Plot title and labels
plt.title('AccountAge vs Churn', fontsize=14)
plt.xlabel('Churn', fontsize=12)
plt.ylabel('Account Age', fontsize=12)

# Update x-axis tick labels
plt.xticks([0, 1], ['Not Churned', 'Churned'])

# Adjust layout
plt.tight_layout()
plt.show()
```

<span style="font-size:16px">The <b>median difference</b> of <b>65 vs 40</b> suggests that customers who have not churned tend to have a higher Account Age compared to those who have churned. This could indicate that customers who have been with the company longer are <u>less likely to churn</u>.</span>


### <u>MonthlyCharges vs. Churn:</u>

**MonthlyCharges (0.10):** A weak positive correlation hints at customers with higher monthly charges being slightly more likely to churn.

```python
plt.figure(figsize=(8, 6))

# Plot the KDE for Churned vs Not Churned
sns.kdeplot(data=churn[churn['Churn'] == 0], x='MonthlyCharges', fill=True, color='skyblue', label='Not Churned', alpha=0.5)
sns.kdeplot(data=churn[churn['Churn'] == 1], x='MonthlyCharges', fill=True, color='lightcoral', label='Churned', alpha=0.5)

# Add labels and title
plt.title('MonthlyCharges vs Churn (KDE Plot)', fontsize=14)
plt.xlabel('Monthly Charges', fontsize=12)
plt.ylabel('Density', fontsize=12)
plt.legend(title='Churn', fontsize=12)
plt.tight_layout()
plt.show()
```

<span style="font-size:16px">Customers with higher monthly charges are <strong>more likely to churn</strong>. This suggests that, as the monthly charges increase, the likelihood of a customer leaving the service increases as well.</span>
